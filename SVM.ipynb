```python
from pandas import Series, DataFrame
```

```python
social = pd.read_csv('Social_Network_Ads.csv')
```

```python
social.head()
```

```python
x = social[['EstimatedSalary','Age']].values
```

```python
y = social['Purchased'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.20, random_state = 42)
```

```python
xtrain
```

```python
xtrain = xtrain.astype(float)
xtest = xtest.astype(float)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
xtrain = sc.fit_transform(xtrain)     #fit store the value in backend
xtest = sc.transform(xtest)
```

```python
from sklearn.svm import SVC       #support vector classifier   and SVM for linear regression
```

```python
classifier = SVC(kernel='rbf', random_state=0)    #instead of linear u can use kernel
classifier.fit(xtrain, ytrain)
```

```python
ypred = classifier.predict(xtest)
```

```python
ypred
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(ytest,ypred)
```
