# Merge

```python
df7 = pd.DataFrame({'ds1' : [0,1,2,3,4,5], 'key1' : ['x','z','y','z','x','x']})
```

```python
df7
```

```python
df8 = pd.DataFrame({'ds2' : [1,2,3], 'key2' : ['q','y','z']})
```

```python
df8
```

```python
df9 = pd.merge(df7,df8,left_on='key1',right_on='key2')
df9
```

```python
df10 = pd.merge(df7,df8,left_on='key1',right_on='key2',how='outer')
df10
```

```python
df11 = pd.merge(df7,df8,left_on='key1',right_on='key2',how='left')
df11    #left side ki puri value aur right side ki common values
```

```python
df12 = pd.merge(df7,df8,left_on='key1',right_on='key2',how='right')
df12   #right side ki puri value aur left side ki common values
```

# duplicated function

duplicated function : 
    check the first value which is unique gives false and check the down value and if the upper is same the gives true and 
    if the upper is not same then false

```python
fg.duplicated() 
```

```python
fg.drop_duplicates()
```

```python
fg['key1'].duplicated()
```

```python
fg.drop_duplicates(['key1'])
```

```python
fg.drop_duplicates(['key1'],keep = 'last') #it check from down to upward
```

```python
fg['key1'].value_counts() #it gives total count of the value of specific column
```
