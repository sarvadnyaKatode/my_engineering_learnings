```python
x=data[['Age','EstimatedSalary']].values ##variable
```

```python
y=data['Purchased'].values   ##label variable
```

```python
from sklearn.model_selection import train_test_split   
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
X_train
```

```python
X_train=X_train.astype(float)
X_test=X_test.astype(float)     ##for specific datatype
```

```python
X_train
```

# feature scalling

```python
#feature scalling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train= sc.fit_transform(X_train)      ## .fit store the values in backend
X_test =sc.transform(X_test)     
#standard scaler range is -3 to +3
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model=LogisticRegression()  #call your algorithm
```

```python
log_model.fit(X_train,Y_train)
```

```python
y_pred=log_model.predict(X_test)
```

```python
y_pred  #predicted
```

```python
Y_test   #Actual
```

```python
from sklearn.metrics import accuracy_score
```

```python
score=accuracy_score(Y_test,y_pred)
```

```python
score*100
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(Y_test,y_pred)
```

```python
cm
```

```python
import seaborn as sns
```

```python
sns.heatmap(cm,annot=True)
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,y_pred))
```
